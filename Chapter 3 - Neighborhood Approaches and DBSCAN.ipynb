# Evaluating the Impact of Neighborhood Radius Size

```python
# import packages
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# create random dataset
X, y = make_blobs(n_samples=1000, centers=8, n_features=2, random_state=800)
```

```python
# visualize the dataset
plt.figure(figsize=(7,5))
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# instantiating DBSCAN
db = DBSCAN(eps=0.5, # this is the neighbor radius
            min_samples=10, # this is the minimum number of points
            metric='euclidean')
```

Let's set up a loop that allows us to explore potential neighborhood radius size options interactively

```python
eps = [0.2, 0.7, 4]

for ep in eps:
    db = DBSCAN(eps=ep, min_samples=10, metric='euclidean')
    db_clusters = db.fit_predict(X)
    plt.scatter(X[:, 0], X[:, 1], c=db_clusters)
    plt.title('DBSCAN with Radius = ' + str(ep))
    plt.show()
```

```python
eps = [0.5, 0.7, 1.3, 1.7]

for ep in eps:
    db = DBSCAN(eps=ep, min_samples=10, metric='euclidean')
    db_clusters = db.fit_predict(X)
    plt.scatter(X[:, 0], X[:, 1], c=db_clusters)
    plt.title('DBSCAN with Radius = ' + str(ep))
    plt.show()
```

# Evaluating the Impact of the Minimum Points Threshold

```python
# create a dataset
X, y = make_blobs(n_samples=1000, n_features=2, centers=8, random_state=800)
```

```python
# visualize the data
plt.figure(figsize=(7,5))
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# instantiate the DBSCAN
db = DBSCAN(eps=0.7, min_samples=10, metric='euclidean')
```

```python
# testing diferent values for min_samples
num_samples = [10, 19, 20]

for min_num in num_samples:
    db = DBSCAN(eps=0.7, min_samples=min_num, metric='euclidean')
    db_clusters = db.fit_predict(X)
    plt.scatter(X[:, 0], X[:, 1], c=db_clusters)
    plt.title('Minimum points :' + str(min_num))
    plt.show()
```
