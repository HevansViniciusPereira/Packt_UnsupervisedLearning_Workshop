# Comparing k-means with Hierarchical Clustering

<b> Import the necessary packages from scikit-learn (KMeans, AgglomerativeClustering, and silhouette_score) </b>

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

<b> Read the wine dataset into a pandas DataFrame and print a small sample </b>

```python
wine = pd.read_csv('datasets/wine_data.csv')
wine.head()
```

<b> Visualize some features from the dataset by plotting the OD Reading feature against the proline feature </b>

```python
plt.figure(figsize=(7, 5))
plt.title('Wine Data')
plt.xlabel('OD Reading')
plt.ylabel('Proline')
plt.scatter(wine['OD_read'], wine['Proline'])
plt.show()
```

<b> Use the sklearn implementation of k-means on the wine dataset, knowing that there are three wine types </b>

```python
km = KMeans(3)
km_clusters = km.fit_predict(wine)
```

```python
km_clusters
```

<b> Use the sklearn implementation of hierarchical clustering on the wine dataset </b>

```python
ac = AgglomerativeClustering(3, linkage='average')
ac_clusters = ac.fit_predict(wine)
```

```python
ac_clusters
```

<b> Plot the predicted clusters from k-means </b>

```python
plt.figure(figsize=(7,5))
plt.scatter(wine.iloc[:, 0], wine.iloc[:, 1], c=km_clusters)
plt.title("Wine Clusters from KMeans Clustering")
plt.xlabel("OD Reading")
plt.ylabel("Proline")
plt.show()
```

<b> Plot the predicted clusters from hierarchical clustering </b>

```python
plt.figure(figsize=(7,5))
plt.title("Wine Clustering from Agglomerative Clustering")
plt.xlabel("OD Reading")
plt.ylabel("Proline")
plt.scatter(wine.iloc[:, 0], wine.iloc[:, 1], c=ac_clusters)
plt.show()
```

<b> Compare the silhouette score of each clustering method </b>

```python
print("Silhouette Score for Wine Dataset:")
print("KMeans Clustering: ", silhouette_score(wine, km_clusters))
print("Agglomerative Clustering: ", silhouette_score(wine, ac_clusters))
```
